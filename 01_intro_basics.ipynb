```python
import matplotlib.pyplot as plt

# Valores de ejemplo de los canales RGB
red_values = [255, 0, 128, 50, 200]
green_values = [0, 255, 64, 150, 100]
blue_values = [0, 128, 255, 200, 50]

# Crear una figura y ejes
fig, ax = plt.subplots()

# Graficar los valores RGB
ax.plot(red_values, label='Rojo')
ax.plot(green_values, label='Verde')
ax.plot(blue_values, label='Azul')

# Etiquetas y título
ax.set_xlabel('Píxeles')
ax.set_ylabel('Valor')
ax.set_title('Valores RGB en una Imagen')

# Agregar una leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
